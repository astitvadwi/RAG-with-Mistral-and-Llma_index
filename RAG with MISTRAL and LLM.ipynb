```python
import logging
import sys

logging.basicConfig(stream=sys.stdout, level=logging.INFO) # Directs log messages to standard output (sys.stdout), so all log messages of level INFO or higher will be printed to the console.
logging.getLogger().addHandler(logging.StreamHandler(stream=sys.stdout)) # Adds a stream handler that also directs log messages to standard output. This can ensure logs are consistently written to the console.

from llama_index.core import VectorStoreIndex,SimpleDirectoryReader,ServiceContext
# VectorStoreIndex: Likely refers to an index that stores vectors (possibly embeddings) for fast lookup, retrieval, or search.
# SimpleDirectoryReader: This class may read files from a directory and likely handles the ingestion of documents from local storage.
# ServiceContext: This might represent a context or environment that manages service-level settings or resources (e.g., models, configurations) when building or querying the index.
```

```python
# medical doc source: https://www.versusarthritis.org/media/24901/fibromyalgia-information-booklet-july2021.pdf
```

```python
documents = SimpleDirectoryReader("/content/").load_data() # we need to give an entire directory to scan not a file so here we are scanning from content
```

```python
import torch
from llama_index.llms.llama_cpp import LlamaCPP #  responsible for loading and running the model based on the LLaMA (Large Language Model Meta AI) architecture using CPU or GPU.
from llama_index.llms.llama_cpp.llama_utils import (messages_to_prompt,completion_to_prompt,) # Utility functions that transform input messages or model outputs into a specific format understood by the LLaMA model.

llm = LlamaCPP(
    # You can pass in the URL to a GGML model to download it automatically, and this is not full mistral-7b it's a quantised iteration ie.., made to run with small gpu's
    model_url='https://huggingface.co/TheBloke/Mistral-7B-Instruct-v0.1-GGUF/resolve/main/mistral-7b-instruct-v0.1.Q4_K_M.gguf',
    # optionally, you can set the path to a pre-downloaded model instead of model_url
    model_path=None,
    temperature=0.1,
    max_new_tokens=256,
    # llama2 has a context window of 4096 tokens, but we set it lower to allow for some wiggle room
    context_window=3900,
    # kwargs to pass to __call__()
    generate_kwargs={},
    # kwargs to pass to __init__()
    # set to at least 1 to use GPU
#    model_kwargs={"n_gpu_layers": 0} # put it value as -1 when we have gpu
    # transform inputs into Llama2 format
    messages_to_prompt=messages_to_prompt,
    completion_to_prompt=completion_to_prompt,
    verbose=True,
)
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.embeddings.langchain import LangchainEmbedding
from llama_index.core.service_context import ServiceContext

embed_model = LangchainEmbedding(
  HuggingFaceEmbeddings(model_name="thenlper/gte-large") # pre-trained model that converts text into embeddings (dense vector representations). These embeddings are typically used for tasks like semantic search, clustering, or similarity comparison between text data.
  # then1per is better embedding format than sentence embedding
)
```

```python
from llama_index.core import Settings #  to configure global settings for the LlamaIndex system

# Set global settings
Settings.chunk_size = 256  # Adjust the chunk size as needed for chunker
Settings.llm = llm   # Assuming 'llm' is defined elsewhere
Settings.embed_model = embed_model  # embbeding model to work with our llms
```

```python
index = VectorStoreIndex.from_documents(documents) # it manages the creation of a vector store. A vector store allows documents to be stored as embeddings (dense vector representations), enabling fast, similarity-based searches on those embeddings.
```

Ask 1 Question

```python
query_engine = index.as_query_engine() # converts the previously created VectorStoreIndex into a query engine.
response = query_engine.query("What is Fibromyalgia?") # The query engine converts the question into a vector (embedding) using the same embedding model that was used to index the documents.
                                                       # It then compares this query vector with the document vectors stored in the index to find the most similar (semantically relevant) documents.
                                                      # The query engine retrieves the most relevant documents from the index and generates a response based on the content found in those documents.
```

```python
print(response) # withour print it prints mostly meta-data with resonse in between
```

Asking Continuos Question

```python
while True:
  query=input()
  response = query_engine.query(query)
  print(response)
```
